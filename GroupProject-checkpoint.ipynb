```python
import matplotlib.pyplot as plt
import pandas as pd

#Reading the dataset
#https://gco.iarc.fr/causes/alcohol/tools-bars?mode=1&sex=0&population=900&population_group=4&country=4&continent=0&cancer=40&key=attr_cases&lock_scale=0&nb_results=10&age_group=3&colored=1
cancer_alc_type = pd.read_csv('C:/Users/oldri/Desktop/CS2410/datasets/CancerAlcohol.csv')
cancer_alc_race = pd.read_csv("C:\\Users\\oldri\\Desktop\\CS2410\\datasets\\CancerAttributable.csv")

#Reading the first file
cases2 = cancer_alc_type["Number of attributable cases"]
cancer2 = cancer_alc_type["Cancer site"]
cases = []
cancer = []

#Turning all the String values into an array of float and making the cancer list one shorter
for i in range(len(cases2)-1):
    cases.append(float(cases2[i]))
    cancer.append(cancer2[i])
print(cases)
#Creating the piechart
plt.pie(cases , labels = cancer, wedgeprops= {'edgecolor': 'black'},autopct = '%1.1f%%') 
plt.title ("Cancers caused by alcohol") 
plt.tight_layout() 
plt.show()

#Reading the second file
also_cases2 = cancer_alc_race["Number of attributable cases"]
races2 = cancer_alc_race["Population"]
also_cases = []
races = []
explode = [0,0,0,0,.5,.5]

#Turning all the String values into an array of float and making the race list one shorter
for i in range(len(also_cases2)-1):
    also_cases.append(float(also_cases2[i]))
    races.append(races2[i])
print(races)
print(also_cases)
#Creating the piechart
plt.pie(also_cases , labels = races, wedgeprops= {'edgecolor': 'black'},autopct = '%1.1f%%',explode = explode) 
plt.title ("Cancers by race") 
plt.tight_layout() 
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

#Percentage by race
#https://gco.iarc.fr/causes/alcohol/tools-multi-bars?mode=1&sex=0&population=900&population_group=4&country=4&continent=0&cancer=40&key=paf&lock_scale=0&nb_results=20&age_group=3&stacked=expanded
df = pd.read_csv("C:\\Users\\oldri\\Desktop\\CS2410\\datasets\\Percentages.csv")

#Creating the necessary lists
moderate2 = df["Moderate (<20 g/day)"]
risky2 = df["Risky (20-60 g/day)"]
heavy2 = df["Heavy (>60 g/day)"]
moderate, risky, heavy = [], [], []
countries = ["Africa","Asia","Europe","Latin America and the Carribean","North America","Oceania"]

#Removing the last index for everything
for i in range(len(moderate2)-1):
    moderate.append(float(moderate2[i].replace("%", "")))
    risky.append(float(risky2[i].replace("%", "")))
    heavy.append(float(heavy2[i].replace("%", "")))

#Moderate bar graph
xs = range(len(moderate))
plt.barh(xs, moderate, color = 'purple', height = .5)
plt.yticks(xs, countries, size = 'small')
plt.title("% of 'Moderate' drinkers by country that got diagnosed with cancer")
plt.show()

#Risky bar graph
plt.barh(xs,risky, color = 'blue', height = .5)
plt.yticks(xs, countries, size = 'small')
plt.title("% of 'Risky' drinkers by country that got diagnosed with cancer")
plt.show()

#Heavy bar graph
plt.barh(xs,heavy, color = 'red', height = .5)
plt.yticks(xs, countries, size = 'small')

plt.title("% of 'Heavy' drinkers by country that got diagnosed with cancer")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
#Sorting the countries by continent
africa ="""Algeria
Angola
Benin
Botswana
Burkina Faso
Burundi
Cabo Verde/Cape Verde
Cameroon
Central African Republic
Chad
Comoros
Congo/Republic of the Congo
Democratic Republic of the Congo
Djibouti
Egypt
Equatorial Guinea
Eritrea
Eswatini
Ethiopia
Gabon
Gambia, The
Ghana
Guinea
Guinea-Bissau
Ivory Coast/Republic of Côte d'Ivoire
Kenya
Lesotho
Liberia
Libya
Madagascar
Malawi
Mali
Mauritania
Mauritius
Morocco
Mozambique
Namibia
Niger
Nigeria
Rwanda
Sao Tome and Principe
Senegal
Seychelles
Sierra Leone
Somalia
South Africa
South Sudan
Sudan
Tanzania
Togo
Tunisia
Uganda
Zambia
Zimbabwe
Gambia
Ivory Coast
DR Congo
"""
africa_list = africa.split('\n')

asia = """Afghanistan
Armenia
Azerbaijan
Bahrain
Timor-Leste
Bangladesh
Bhutan
British Indian Ocean Territory
Brunei
Cambodia
China
Cyprus
Egypt
Georgia
Hong Kong
India
Indonesia
Iran
Iraq
Israel
Japan
Jordan
Kazakhstan
Kuwait
Kyrgyzstan
Laos
Lebanon
Macau
Malaysia
Maldives
Mongolia
Myanmar
Nepal
North Korea
Oman
Pakistan
Palestine
Philippines
Qatar
Russia
Saudi Arabia
Singapore
South Korea
Sri Lanka
Syria
Taiwan
Tajikistan
Thailand
Timor-Leste/East Timor
Turkey
Turkmenistan
United Arab Emirates
Uzbekistan
Vietnam
Yemen
"""
asia_list = asia.split('\n')

europe ="""Albania
Andorra
Armenia
Austria
Azerbaijan
Belarus
Belgium
Bosnia and Herzegovina
Bulgaria
Croatia
Cyprus
Czechia/Czech Republic
Denmark
Estonia
Finland
France
Georgia
Germany
Greece
Hungary
Iceland
Ireland
Italy
Kazakhstan
Latvia
Liechtenstein
Lithuania
Luxembourg
Malta
Moldova
Monaco
Montenegro
Netherlands
North Macedonia
Norway
Poland
Portugal
Romania
Russia
San Marino
Serbia
Slovakia
Slovenia
Spain
Sweden
Switzerland
Turkey
Ukraine
United Kingdom
Vatican City
"""
europe_list = europe.split('\n')

na = """Antigua and Barbuda
Bahamas
Barbados
Belize
Canada
Costa Rica
Cuba
Dominica
Dominican Republic
El Salvador
Grenada
Guatemala
Haiti
Honduras
Jamaica
Mexico
Nicaragua
Panama
Saint Kitts and Nevis
Saint Lucia
Saint Vincent and the Grenadines
Trinidad and Tobago
United States
urmom
"""
na_list = na.split()

oceania = """Australia
Fiji
Kiribati
Marshall Islands
Micronesia (The Federated States of)
Nauru
New Zealand
Palau
Papua New Guinea
Samoa
Solomon Islands
Tonga
Tuvalu
Vanuatu
Cook Islands
Niue
Micronesia
"""

oceania_list = oceania.split('\n')

Latin_America ="""Argentina
Bolivia
Brazil
Chile
Colombia
Ecuador
Guyana
Paraguay
Peru
Suriname
Uruguay
Venezuela
Antigua and Barbuda
Dominican Republic
Trinidad and Tobago
Saint Kitts and Nevis
El Salvador
Costa Rica
Saint Lucia
Saint Vincent and the Grenadines

"""

latin_america_list = Latin_America.split('\n')

#Function that checks which continent the country is in
def continent_checker(country):
    if country in africa_list:
        return "Africa"
    elif country in asia_list:
        return "Asia"
    elif country in europe_list:
        return "Europe"
    elif country in na_list:
        return "North America"
    elif country in oceania_list:
        return "Oceania"
    elif country in latin_america_list:
        return "Latin America and The Carribean"
    else:
        return "None"

    
#initializing data
df = pd.read_csv("C:\\Users\\oldri\\Desktop\\CS2410\\datasets\\AlcoholWorld.csv")
countries = df['country']
rate = df['both']
continent = []

#making a list of the countries by continent
for word in countries:
    continent.append(continent_checker(word))
    
continent[37] = "North America"

number_countries = {"Africa": 0,
                    "Asia": 0,
                    "Europe": 0,
                    "North America":0,
                    "Oceania": 0,
                    "Latin America and The Carribean": 0}
#Getting the sum of all countries to later divide
africa_total = 0
asia_total = 0
europe_total = 0
na_total = 0
oceania_total = 0
la_total = 0
#Now count how many countries for each
for i in range(len(continent)):
    if continent[i] == "Africa":
        number_countries["Africa"] += 1
        africa_total += rate[i]
    elif continent[i] == "Asia":
        number_countries["Asia"] += 1
        asia_total += rate[i]
    elif continent[i] == "Europe":
        number_countries["Europe"] += 1
        europe_total += rate[i]
    elif continent[i] == "North America":
        number_countries["North America"] += 1
        na_total += rate[i]
    elif continent[i] == "Oceania":
        number_countries["Oceania"] += 1
        oceania_total += rate[i]
    elif continent[i] == "Latin America and The Carribean":
        number_countries["Latin America and The Carribean"] += 1
        la_total += rate[i]

print(number_countries)

averages = [africa_total/number_countries["Africa"],
            asia_total/number_countries["Asia"],
            europe_total/number_countries["Europe"],
            na_total/number_countries["North America"],
            oceania_total/number_countries["Oceania"],
            la_total/number_countries["Latin America and The Carribean"]]
continents = ["Africa","Asia","Europe","North America","Oceania","Latin America and The Carribean"]

print(averages)
xs = range(len(continents))
plt.barh(xs,averages, color = ['purple','blue','orange','green','brown','red'], height = .5)
plt.yticks(xs, continents, size = 'small')
plt.xlabel("Alcohol consumption (liters)")
plt.title("Average consumption of pure alcohol in liters per year among all adults aged 15+ in 2019 (by continent)")
plt.show()
```
